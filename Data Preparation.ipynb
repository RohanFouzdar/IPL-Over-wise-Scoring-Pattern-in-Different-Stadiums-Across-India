```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import matplotlib.pyplot as plt
%matplotlib inline

warnings.filterwarnings('ignore')
```

```python
data1 = pd.read_csv("C:\\Users\\rohan.fouzdar\\Downloads\\all_matches.csv")
data1.head()
```

```python
print(data1.columns)
```

```python
d2015=data1[data1.start_date>="2015-01-01"]
```

```python
d2015['runs_on_ball']=d2015['runs_off_bat']+d2015['extras']
d2015.head()
```

```python
d2015['Over']=d2015['ball'].apply(np.ceil)
```

```python
d2015['venue'].value_counts()
```

```python
d2015['overkey']=d2015['match_id'].astype(str)+d2015['season'].astype(str)+d2015['Over'].astype(str)+d2015['innings'].astype(str)
doverruns=d2015[d2015.innings>0].groupby('overkey')['runs_on_ball'].sum().to_frame().reset_index()

doverruns.head()
```

```python
doverruns.rename(columns= {'runs_on_ball': 'runs_in_over'})
```

```python
d2015=d2015.merge(doverruns,on='overkey')
```

```python
d2015.head()
```

```python
d2015.rename(columns={'runs_on_ball_x': 'runs_on_ball', 'runs_on_ball_y': 'runs_in_over'}, inplace=True)
```

```python
d2015['venue'].unique()
```

```python
d2015['venue']=d2015['venue'].replace(['M.Chinnaswamy Stadium'],'M Chinnaswamy Stadium')
d2015['venue']=d2015['venue'].replace(['Feroz Shah Kotla'],'Arun Jaitley Stadium')
d2015['venue']=d2015['venue'].replace(['MA Chidambaram Stadium, Chepauk','MA Chidambaram Stadium, Chepauk, Chennai'],'MA Chidambaram Stadium')
d2015['venue']=d2015['venue'].replace(['Wankhede Stadium, Mumbai'],'Wankhede Stadium')
d2015['venue']=d2015['venue'].replace(['Rajiv Gandhi International Stadium, Uppal'],'Rajiv Gandhi International Stadium')
d2015['venue']=d2015['venue'].replace(['Punjab Cricket Association IS Bindra Stadium','Punjab Cricket Association IS Bindra Stadium, Mohali'],'Punjab Cricket Association Stadium, Mohali')
```

```python
d2015.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\d2015.csv',sep=',',index=False)
```
