```python
import pandas
 
# reading the CSV file
d2015 = pandas.read_csv('C:\\Users\\rohan.fouzdar\\Downloads\\d2015.csv')
 
# displaying the contents of the CSV file
print(d2015.shape)
```

```python
d2015=d2015[d2015['year']!=2021]
```

```python
d2015['year'].value_counts()
```

```python
dBLR=d2015[d2015.venue=='M Chinnaswamy Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dMCA=d2015[d2015.venue=='Maharashtra Cricket Association Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dDEL=d2015[d2015.venue=='Arun Jaitley Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dKOL=d2015[d2015.venue=='Eden Gardens'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dCHE=d2015[d2015.venue=='MA Chidambaram Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dMUM=d2015[d2015.venue=='Wankhede Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dGUJ=d2015[d2015.venue=='Sardar Patel Stadium, Motera'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dLUC=d2015[d2015.venue=='Green Park'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dRAJ=d2015[d2015.venue=='Sawai Mansingh Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dHYD=d2015[d2015.venue=='Rajiv Gandhi International Stadium'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
dPUN=d2015[d2015.venue=='Punjab Cricket Association Stadium, Mohali'].groupby(["venue","year","Over"])['runs_in_over'].mean().to_frame().reset_index()
```

```python
dBLR1=dBLR.pivot(index='Over',columns='year',values='runs_in_over')
dMCA1=dMCA.pivot(index='Over',columns='year',values='runs_in_over')
dDEL1=dDEL.pivot(index='Over',columns='year',values='runs_in_over')
dKOL1=dKOL.pivot(index='Over',columns='year',values='runs_in_over')
dCHE1=dCHE.pivot(index='Over',columns='year',values='runs_in_over')
dMUM1=dMUM.pivot(index='Over',columns='year',values='runs_in_over')
dGUJ1=dGUJ.pivot(index='Over',columns='year',values='runs_in_over')
dLUC1=dLUC.pivot(index='Over',columns='year',values='runs_in_over')
dRAJ1=dRAJ.pivot(index='Over',columns='year',values='runs_in_over')
dHYD1=dHYD.pivot(index='Over',columns='year',values='runs_in_over')
dPUN1=dPUN.pivot(index='Over',columns='year',values='runs_in_over')
```

```python
dBLR1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dBLR1.csv',sep=',',index=False)
dMCA1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dMCA1.csv',sep=',',index=False)
dDEL1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dDEL1.csv',sep=',',index=False)
dKOL1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dKOL1.csv',sep=',',index=False)
dCHE1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dCHE1.csv',sep=',',index=False)
dMUM1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dMUM1.csv',sep=',',index=False)
dGUJ1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dGUJ1.csv',sep=',',index=False)
dLUC1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dLUC1.csv',sep=',',index=False)
dRAJ1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dRAJ1.csv',sep=',',index=False)
dHYD1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dHYD1.csv',sep=',',index=False)
dPUN1.to_csv('C:\\Users\\rohan.fouzdar\\Downloads\\dPUN1.csv',sep=',',index=False)
```
